```python
# This remove filler from cabinets

def fillerRemoveFunction():
    filler = driver.find_element_by_name('straight-strip-toggle')
    driver.execute_script("arguments[0].scrollIntoView();", filler)
    time.sleep(1)

    filler.click()

    time.sleep(1)
    Frontpath = driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[1]')
    driver.execute_script("arguments[0].scrollIntoView();", Frontpath)



# This function select the front option to display list of fronts
def selectingFrontOption():
    try:
        driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[3]/div/div/button').click()
    except:
        try:
            driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[4]/div/div/button').click()
        except:
            driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[5]/div/div/button[3]').click()



# This function helps us to change the front of the cabinet and return the prices

def pricingFunction():
    front_list = driver.find_element_by_id(f'mod-product-list')
    price = []
            
    fronts = front_list.find_elements_by_tag_name("li")

    pricing_groups = ['244 - Speed (Prog 351)',
    '427 - Laser (Prog 615)',
    '453 - Flash ( Prog 648)',
    '165 - Fashion ( Prog 406)',
    '551 - Inline ( Prog 526)',
    '776 - Cascada ( Prog 684)',
    '390 - Castello ( Prog 224)',
    '819 - Lux ( Prog 555)',
    '938 - Artis ( Prog 789)'
    ]

    referencePggroups = [
        'PG1','PG2','PG3','PG4','PG5','PG7','PG8','PG9','Pg10'
    ]

    frontsNameList = []
    
    for front in fronts:
        frontsNameList.append(front.find_elements_by_tag_name('p')[0].text)


    for pg in pricing_groups:
        if pg in frontsNameList:
            for front in fronts:
                    if front.find_elements_by_tag_name("p")[0].text == pg:
                        front.click()
                        time.sleep(4)
                        priceValue = driver.find_element_by_xpath('/html/body/div[3]/main/div/div[1]/div/div[2]/p/span[2]/span/span/span').text
                        price.append(priceValue)
        else:
            price.append('No Pricing Group')
    
    price.insert(5,'0,00')
    

    return price

# This function remove worktop from base cabinet
    
def deletingWorktop():
    try:
        driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[6]/div/div/button[3]').click()
    except:
        try:
             driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[7]/div/div/button[3]').click()
        except:
            try:
                driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[5]/div/div/button[3]').click()
            except:
                driver.find_element_by_xpath('//*[@id="design-edit-panel"]/div[2]/div/ul/li[4]/div/div/button[3]').click()



# getting final price after removing unwanted spaces and other things

def gettingFinalPrice():
    finalprice = []
    for i in prices:
        for j in i:
            if j!='No price':
                if j!= 'Prix indisponible':
                    if j!= 'Calcul du prix en cours...':
                        a = (j.replace(' â‚¬',''))
                        a = a.replace(' ','')
                        # a = float(a.replace(',','.'))
                        # a = (round(a-130.4,2))
                        finalprice.append(str(a).replace('.',','))
                    else:
                        finalprice.append(j)
                else:
                    finalprice.append(j)
            else:
                finalprice.append(j)
    return finalprice



# Collecting pricing for open cabintes

def getPriceForOpenCabinets():

        prices =[]
        price = driver.find_element_by_xpath('/html/body/div[3]/main/div/div[1]/div/div[2]/p/span[2]/span/span/span').text
        prices.append(price)

        prices = [prices[0] for _ in range(10)]

        return prices


# this function performs cancel operation

def cancelButton():
    try:
        cancel =driver.find_element_by_xpath('//*[@id="header-filter"]/ul/li[2]/button')
        cancel.click()
    except:
        cancel =driver.find_element_by_xpath('//*[@id="design-edit-panel"]/header/ul/li/button')
        cancel.click()






    

```
